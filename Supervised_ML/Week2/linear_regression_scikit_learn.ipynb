## Linear Regression using Scikit-Learn
### Goals

Utilize scikit-learn to implement linear regression using a close form solution based on the normal equation

### Load Libraries and tools

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from lab_utils_multi import load_house_data
plt.style.use('./deeplearning.mplstyle')
np.set_printoptions(precision=2)
```

### Create Dataset

```python
# Create data for house prices, features are size in square feet and price in 1000s
X_train = np.array([1.0, 2.0])
# Reshape to 2D array
X_train = X_train.reshape(-1, 1)

y_train = np.array([300.0, 500.0])
```

### Create and fit the model

```python
Linear_model = LinearRegression()
Linear_model.fit(X_train, y_train)
```

```python
# View parameters/coefficients of the model w and b
w = Linear_model.coef_
b = Linear_model.intercept_

print(f'w: {w}')
print(f'b: {b}')
```

```python
# Make Predictions
y_pred = Linear_model.predict(X_train)

# Print Predictions
print(f'y_pred: {y_pred}')
```

```python
# Make Predictions for new data
X_new = np.array([3.0, 4.0])
X_new = X_new.reshape(-1, 1)
y_new = Linear_model.predict(X_new)

print(f'y_new: {y_new}')
```

### Another example with multiple features

```python
# Load house data
X_train, y_train= load_house_data()
```

```python
# Initialize the model
Linear_model = LinearRegression()

# Fit the model
Linear_model.fit(X_train, y_train)

# View parameters/coefficients of the model w and b
w = Linear_model.coef_
b = Linear_model.intercept_

print(f'w: {w}')
print(f'b: {b}')
```

```python
# Make Predictions
y_pred = Linear_model.predict(X_train)

# Print Predictions
print(f'y_pred: {y_pred[:4]}')
```

```python
# Plot Actual vs Predicted Values
fig, ax = plt.subplots(1, 1, figsize=(8, 6))

# Plot actual values (using a scatter plot)
ax.scatter(range(len(y_train)), y_train, color='blue', marker='o', label='Actual Values')

# Plot predicted values (using a scatter plot)
ax.scatter(range(len(y_pred)), y_pred, color='red', marker='x', label='Predicted Values')

# Add plot labels and title
ax.set_xlabel('Sample Index')
ax.set_ylabel('Values')
ax.set_title('Actual vs Predicted Values')

# Add legend
ax.legend()

# Show plot
plt.show()
```
